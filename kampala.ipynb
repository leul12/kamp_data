### let's `Merge` Them 
to do that you can do it in two ways 
*  give list pf columns to merge them
* idetify the one column that's the same in all file and merge on that

```python
data_frames=[df_feb,df_march,df_april,df_may]
```

```python
df = pd.concat(data_frames,ignore_index=True)
```

# Where it Might be `?`

```python
fig = px.scatter_mapbox(df, lat=df.lat, lon=df.lon, zoom=6, height=300)
fig.update_layout(
    mapbox_style="white-bg",
    mapbox_layers=[
        {
            "below": 'traces',
            "sourcetype": "raster",
            "sourceattribution": "United States Geological Survey",
            "source": [
                "https://basemap.nationalmap.gov/arcgis/rest/services/USGSImageryOnly/MapServer/tile/{z}/{y}/{x}"
            ]
        }
    ])
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()
```

### monthly segmented data

```python
dffeb=df.loc[0:16981, ['timestamp', 'value_type', 'value']]

dfmar=df.loc[16982:54570, ['timestamp', 'value_type', 'value']]

dfapr=df.loc[54572:57293, ['timestamp', 'value_type', 'value']]

dfmay=df.loc[57293:, ['timestamp', 'value_type', 'value']]

```

# let's see the values of humidity and staf during the months

```python
fig = px.line(dffeb, x='timestamp', y='value', color='value_type',title='February Data')
fig.show()
```

```python
fig = px.line(dfmar, x='timestamp', y='value', color='value_type',title='March Data')
fig.show()
```

```python
fig = px.scatter(dfapr, x='timestamp', y='value', color='value_type',size='value' ,title='April Data')
fig.show()
```

```python
fig = px.line(dfmay, x='timestamp', y='value', color='value_type',title='May Data')
fig.show()
```
